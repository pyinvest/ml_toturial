```python
from sklearn.decomposition import PCA
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris=datasets.load_iris()
X=iris.data
y=iris.target
```

```python
pca=PCA(n_components=2)
pca.fit(X).transform(X)
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```
