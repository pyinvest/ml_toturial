```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.DataFrame([['A','windy','hot',25],['B','sunny','hot',30],['B','cloudy','cold',18]],columns=['Area','weather','feeling','temperature'])
df=data.copy()
df
```

# OneHotEncoder

```python
enc=OneHotEncoder()
```

```python
le=LabelEncoder()
for col in df[['Area','weather','feeling']]:
    df[col]=le.fit_transform(df[col])
df
```

```python
df_ohe=enc.fit_transform(df).toarray()
pd.DataFrame(df_ohe)
```

```python
enc_1=OneHotEncoder(categorical_features=[1])
```

```python
df_ohe=enc_1.fit_transform(df).toarray()
pd.DataFrame(df_ohe)
```
